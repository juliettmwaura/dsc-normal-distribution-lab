## A quick refresher! 

Here's the formula for the normal distribution density function once more:

$$ \large N(x) = \dfrac{1}{\sigma \sqrt {2\pi }}e^{-\dfrac{(x-\mu)^2}{2\sigma^2}}$$

Here, 
- $\mu$ is the mean
- $\sigma$ is the standard deviation
- $\pi \approx 3.14159 $ 
- $ e \approx 2.71828 $


## First generate a normal distribution containing 5000 values with $\mu=14$ and $\sigma = 2.8$

```python
# Generate a random normal variable with given parameters , n=5000
import numpy as np

# Parameters for the normal distribution
mu = 14
sigma = 2.8
n = 5000

# Generate random normal variables
normal_values = np.random.normal(mu, sigma, n)

# Display the first 10 values for verification
print(normal_values[:10])
```

## Calculate a normalized histogram for this distribution in matplotlib, with bin size = 20

Make sure to get the bin positions and counts for each of the obtained bins. You can use [official documentation](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.hist.html) to view input and output options for `plt.hist()`

```python
# Calculate a histogram for above data distribution

import matplotlib.pyplot as plt

# Parameters for the normal distribution
mu = 14
sigma = 2.8
n = 5000

# Generate random normal variables
normal_values = np.random.normal(mu, sigma, n)

# Create histogram with specified bins
bin_size = 20
hist, bins, _ = plt.hist(normal_values, bins=bin_size, density=True, alpha=0.7)

# Plot settings
plt.xlabel('Values')
plt.ylabel('Normalized Frequency')
plt.title('Normalized Histogram of Normal Distribution')
plt.grid(True)
plt.show()

# Display bin positions and counts
print("Bin Positions:", bins)
print("Bin Counts:", hist)
```

## Use the formula to calculate the density function with $\mu$, $\sigma$ and bin information obtained before

## Plot the histogram and density function

```python
# Plot histogram along with the density function
import numpy as np
import matplotlib.pyplot as plt

# Parameters for the normal distribution
mu = 14
sigma = 2.8
n = 5000

# Generate random normal variables
normal_values = np.random.normal(mu, sigma, n)

# Create histogram with specified bins
bin_size = 20
hist, bins, _ = plt.hist(normal_values, bins=bin_size, density=True, alpha=0.7)

# Calculate bin centers
bin_centers = (bins[:-1] + bins[1:]) / 2
# Calculate the normal Density function 
# Calculate bin centers
bin_centers = (bins[:-1] + bins[1:]) / 2

# Calculate PDF for each bin using the normal distribution formula
density = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-((bin_centers - mu)**2) / (2 * sigma**2))

# Plot settings
plt.xlabel('Values')
plt.ylabel('Normalized Frequency / Density')
plt.title('Normalized Histogram and Density Function of Normal Distribution')
plt.grid(True)
```

## Visualize the distribution using seaborn and plot the KDE

```python
# Use seaborn to plot the histogram with KDE
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Parameters for the normal distribution
mu = 14
sigma = 2.8
n = 5000

# Generate random normal variables
normal_values = np.random.normal(mu, sigma, n)

# Create seaborn plot with histogram and KDE
sns.histplot(normal_values, kde=True, bins=20, color='blue', stat='density')

# Plot settings
plt.xlabel('Values')
plt.ylabel('Density')
plt.title('Distribution of Normal Random Variables with KDE')
plt.grid(True)

plt.show()
```
